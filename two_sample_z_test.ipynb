```python
# H0: p1 = p2
# HA: p1 != p2

# generate data
n1 = 100    # number of trials in sample
n2 = 80
k1 = 25     # number of successes in sample
k2 = 20
x1 = k1/n1  # success rate of sample
x2 = k2/n2
```

```python
# implement 2-sample z-test
Xpool = (k1+k2) / (n1+n2) 
sigma_hat = np.sqrt( Xpool*(1- Xpool) )
z = (x1 - x2) / ( sigma_hat*np.sqrt( n1**(-1) + n2**(-1) ) )


# compute p-values
def z_test(z_stat):
    from scipy.stats import norm    
	# calculate the p-value
    p = 1 - norm.cdf(abs(z_stat))               # one sided test
    return p

print(z_test(z))



# used for e.g. hetero-/homogeneity in inter-/intragrade testing                       # check normalization; in real life there is termination, so sum is not equal to 1
```
