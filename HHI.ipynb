## 1. Herfindahl-Hirschman Index (HHI)

```python
# HHI function
def HHI(n_ppl):
    tot_ppl = np.sum(n_ppl)
    share = n_ppl / tot_ppl
    # compute HHI
    HHI = np.sum(share**2)
    return HHI

# number of ppl per bucket
n_ppl = np.round(np.random.rand(20)*10, 0)
print(HHI(n_ppl))

# let's look at some extreme cases
# complete equality
n_ppl = np.ones(2)
print(HHI(n_ppl))

n_ppl = np.ones(10000000) # HHI also sensitive for n_tot
print(HHI(n_ppl))

# complete inequality
n_ppl = np.zeros(20)
n_ppl[3] = 10
n_ppl[18] = 100
print(HHI(n_ppl))

# or actually:
n_ppl = np.zeros(20)
n_ppl[18] = 10000
print(HHI(n_ppl))


# note:
# one can prove that Gini <= (n_tot**2 - HHI) / (n_tot**2 - n_tot)
# which for large N converges to: g <= 1 - HHI
# this shows that a low Gini may not be due to low discriminatory power, but to inadequate bucketing


```

## 2. Population Stability Index (PSI)

```python
############################################
# PSI
# code currently not robust against 0 ppl in bucket
# proper implementation shld do smth like assigning at least 5% of obs to a bucket
# idea stays the same though
def PSI(n_ppl_old, n_ppl_new):              
    tot_ppl_new = np.sum(n_ppl_new)
    tot_ppl_old = np.sum(n_ppl_old)
    share_new = n_ppl_new / tot_ppl_new
    share_old = n_ppl_old / tot_ppl_old
    # compute PSI
    PSI = np.sum( (share_old - share_new) * (np.log(share_old) - np.log(share_new)) )    
    return PSI

# number of ppl per bucket + small change
n_ppl_old = np.round(np.random.rand(20)*10, 0)
n_ppl_new = n_ppl_old +  np.round(np.random.rand(20), 0) - np.round(np.random.rand(20), 0)
print(PSI(n_ppl_old, n_ppl_new))

# big change
n_ppl_old = np.round(np.random.rand(20)*10, 0)
n_ppl_new = n_ppl_old +  np.round(np.random.rand(20)*20, 0) 
print(PSI(n_ppl_old, n_ppl_new))

# note from quora:
# If PSI<= 0.1 indicates little change [no action required]
# 0.1 - 0.25 is little change but to small to determine [still no action required]  # I doubt this seeing the example above ...
# PSI> 0.25 is a significant shift [action required - merits further investigation]

# comment client documentation
# psi not well defined if either of the shares equals 0. Could be resolved by not summing over empty classes,
# but this is bad practice if share_new is not equal to share old.
```
